Data recinnaissance du locuteur (chirac/ Mitterrand)

```python

# Chargement des données:
def load_pres(fname):
    alltxts = []
    alllabs = []
    s=codecs.open(fname, 'r','utf-8') # pr en lecture et urf-8 pour les caractérees spéciaux

    while True:
        txt = s.readline() #lire tout le texte 
        if(len(txt))<5:
            break
        #
        lab = re.sub(r"<[0-9]*:[0-9]*:(.)>.*","\\1",txt)
        txt = re.sub(r"<[0-9]*:[0-9]*:.>(.*)","\\1",txt)
        if lab.count('M') >0:
            alllabs.append(-1)
        else:
            alllabs.append(1)
        alltxts.append(txt)
    return alltxts,alllabs

```

```python
fname = "./Dataset/corpus.tache1.learn.utf8"
alltxts,alllabs = load_pres(fname)
```

Data pour les film 


```python
def load_movies(path2data): # 1 classe par répertoire
    alltxts_f = [] # init vide
    labs = []
    cpt = 0
    for cl in os.listdir(path2data): # parcours des fichiers d'un répertoire
        for f in os.listdir(path2data+cl):
            txt = open(path2data+cl+'/'+f).read()
            alltxts_f.append(txt)
            labs.append(cpt)
        cpt+=1 # chg répertoire = cht classe

    return alltxts,labs
```

fonction pour clean  the data 


```python
import re
import nltk
import unicodedata
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
stemmer = PorterStemmer ()

# Télécharger les mots d'arrêt 
nltk.download('stopwords')

def clear_text(text, accents=True, minscules=True, ponctuali=True, chifre=True, mot_dar=True,stemming=True ):
    # 1. Supprimer les accents si demandé
    if accents:
        text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode('utf-8')
    
    # 2. Mettre en minuscules si demandé
    if minscules:
        text = text.lower()
    
    # 3. Supprimer la ponctuation si demandé
    if ponctuali:
        text = re.sub(r'[^\w\s]', '', text)
    
    # 4. Supprimer les chiffres si demandé
    if chifre:
        text = re.sub(r'\d+', '', text)
    
    # 5. Supprimer les mots d'arrêt si demandé
    if mot_dar:
        stop_words = stopwords.words('english') + stopwords.words('french')
        text = ' '.join([word for word in text.split() if word not in stop_words])
    if stemming:
        
        stemmer = PorterStemmer ()
        text=' '.join([stemmer.stem(word) for word in text.split()])


    return text
```
